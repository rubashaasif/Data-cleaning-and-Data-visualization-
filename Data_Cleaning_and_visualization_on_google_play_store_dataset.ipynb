```python
df['App'] =df['App'].str.strip()
```

```python
df['Category'].unique()
```

```python
df['Category']= df['Category'].str.strip()
```

```python
df['Category'].isnull().sum()
```

```python
df['Rating'].unique()
```

```python
#checking outliers
sns.boxplot(x=df['Rating'])
plt.title('checking outliers')
plt.xlabel('Rating')
plt.show()
```

```python
df['Rating'].isnull().sum()
```

```python
#removing outlier, ratings more than 5
df.loc[df['Rating'] > 5, 'Rating'] =np.nan
```

```python
#filling missing values with the median , because the rating column has outliers, so it less affected by the median  , and we usually use median for the ratings
df['Rating']= df['Rating'].fillna(df['Rating'].median())
df['Rating'].isnull().sum()
```

```python
df['Reviews'].unique()
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df['Reviews'].isnull().sum()
```

```python
df = df.dropna(subset=['Reviews']) #dropped null value because there was only one
df['Reviews'].isnull().sum()
```

```python
df['Size'].unique()
```

```python
#value are in k (kilobytes ) and MB (megabytes), so i converted all in MB
def value_conversion(x):
    if pd.isna(x):
        return np.nan
    elif 'k' in str(x):
        return float(x.replace('k', '')) / 1024
    elif 'M' in str(x):
        return float(x.replace('M', ''))
df['Size'] = df['Size'].apply(value_conversion)
```

```python
df['Size'].unique()
```

```python
df['Size'] =pd.to_numeric(df['Size'], errors= 'coerce')
```

```python
df['Size'].isnull().sum()
```

```python
df['Size'] = df['Size'].fillna(df['Size'].median())
```

```python
sns.boxplot(x=df['Size'])
plt.show()
```

```python
df['Type'].unique()
```

```python
df['Type'].isnull().sum()
```

```python
df = df.dropna(subset=['Type'])
df['Type'].isnull().sum()
```

```python
df['Installs'].unique()
```

```python
df['Installs'] = df['Installs'].str.replace(',','', regex=False)
df['Installs'] = df['Installs'].str.replace('+','', regex=False)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
df['Installs'].unique()
```

```python
df['Installs'].isnull().sum()
```

```python
df['Price'].unique()
```

```python
df['Price'] = df['Price'].str.replace('$','', regex=False)
df['Price'] = df['Price'].str.strip()
df['Price'].unique()
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Price'].unique()
```

```python
df['Price'].isnull().sum()
```

```python
df['Content Rating'].unique()
```

```python
df['Content Rating'].isnull().sum()
```

```python
df['Genres'].unique()
```

```python
df['Genres'].isnull().sum()
```

```python
df['Last Updated'].unique()
```

```python
df['Last Updated'] =pd.to_datetime(df['Last Updated'], dayfirst=True, errors='coerce' )
df['Last Updated'].unique()
```

#Data Transformation

```python
#values were too many , so grouped in quartiles
df['Installs'] = pd.qcut(df['Installs'], 4, labels= ['low', 'medium', 'high', 'very high'])
```

```python
sns.histplot(df['Installs'],palette='colorblind', edgecolor= 'green')
plt.title('Distribution of Installs')
plt.xlabel('Installs', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.show()
```
